## Machine learning in astrophysics

As a motivation, here are a few examples of problems where machine learning techniques are of great help. This list is not exhaustive and rather biased. 

### Astrophysics:

* **Exoplanet Detection**: ML can assist in the identification of exoplanets in light curves but also in coronographic data. 
* **Cosmic Ray Detection**: Identifying cosmic ray events in astronomical images is a challenging task that machine learning can help with, improving the accuracy and efficiency of data analysis.
* **Lensed quasar searches**: Gravitationally lensed objects are pretty rare objects. Machine learning can be used to find them in big data sets, and even to perform automated modeling of those systems. 
* **Photometric Redshift Estimation**: Predicting the redshift of distant galaxies based on mutlicolor information is a difficult taks. Machine learning techniques are being developped to tackle this problems for the billions of galaxies that are being observed with large scale sky surveys (Euclid, PanStarr, Legacy Survey, Large Synoptic Sky Telescope, ...). 
* **Gravitational Wave Detection**: ML algorithms can aid in the detection and characterization of gravitational waves, as seen in experiments like LIGO and Virgo.
* **Stellar Classification**: Automated classification of stars based on their spectra using machine learning can contribute to understanding stellar populations and evolution.
* **Galaxy Classification**: Machine learning algorithms can be used to automatically classify galaxies based on their morphology, helping astronomers analyze large datasets from surveys like the Sloan Digital Sky Survey (SDSS).


### Geophysics:

* **Earthquake Prediction**: Machine learning models can analyze seismic data to identify patterns and precursor signals that may help in predicting earthquakes and mitigating their impact.
* **Subsurface Imaging**: ML algorithms can enhance the interpretation of geophysical surveys, such as seismic or ground-penetrating radar data, improving the resolution and accuracy of subsurface imaging.
* **Climate Modeling**: Machine learning can be applied to climate data to improve the accuracy of climate models, aiding in predicting long-term climate trends and understanding climate change impacts.
* **Remote Sensing for Environmental Monitoring**: ML techniques can analyze satellite imagery and other remote sensing data to monitor changes in land cover, deforestation, and other environmental parameters.

## Supervised and unsupervised learning 

### Supervised learning: 

This is the task of learning a function that maps input attributes to an outcome value based on a set of **training** examples. 

What does it mean? In general terms this means that you make predictions about "something" (a discrete value, a continuous value or a label) based on a model that you have applied to a given data set, generally called your **training sample**. 

In that framework, the "regression" problem (see SPAT0002-1) enters that category, and is a very basic (but important) machine learning task. Like [Mr Jourdain](https://en.wikipedia.org/wiki/Le_Bourgeois_gentilhomme) who was doing prosody without knowing, you have been doing machine learning "without knowing". Machine learning effectively strongly benefits / makes a strong usage of this information encompassed in the data. The theoretical fundation of the field are statistics and therefore a good understanding of statistics is required to make a sensible use of machine learning machinery or to develop new machine learning algorithms.

### Unsupervised learning 

Supervised learning learning requires a data set for which the data have been labeled / tagged by an expert. The training is based on this external knowledge given to the machine. For unsuperivsed learning, patterns are learned from unlabeled data. The hope is that the computer will create a statistically meaningful representation of the world based on data by identifying structures. These unsupervised algorithm are generally based on some kind of artificial neural networks but beware, that artificial neural network (ANN) does not imply unsupervised learning. ANN are simply a class of machine learning algorithm. They can be trained on labeled data, or used in the framework of unsupervised tasks. The "deep blue" machine that has beaten the Chess Master Gary Kasaparov had been trained on a large variety of "chess games" (to some extent deep blue mostly benefited from a big database rather than advanced machine learning). This was a supervised algorithm. Later on, other algorithm have been devised, where the computer was "playing agains himself" knowing how to move chess pieces, or playing Go, as in the case of "Alpha Go" (and its subsequent versions). In that case the learning is unsupervised. 

### Summary 

- Supervised learning: Models that can predict labels based on labeled training data
- Classification: Models that predict labels as two or more discrete categories
- Regression: Models that predict continuous labels
- Unsupervised learning:  Models that identify structure in unlabeled data
- Clustering: Models that detect and identify distinct groups (clusters) in the data
- Dimensionality reduction: Models that detect and identify lower-dimensional structure in higher dimensional data

### First step with scikit-learn

Before we start, we need to introduce an important aspect machine learning in terms of data management: *don't "eat" all your data at once*. Because machine learning are  based on large amounts of data, you can work on the statistical information that is imprinted in the data themselves!  

In practice, the machine learning task generally consists of taking a set of $n$ samples of data, and then *predict* the properties of unknown data. The approach that consists in deriving the parameters of the model based on a data set is called **training**. But beware, one never trains a model on a full data set. Instead, one splits the data sets into two or 3 subsets: 
- A **training** set: typically $\sim$ 50-70% of the sample. This is the sample on which the model is first train. 
- A **(cross)-validation** set: typically half of the remaining sample. We'll see below the usefulness of this subsample. 
- A **test set**: The other half of the remaining sample. This is a sample "unseen by the model" on which the model efficacy is tested. 

At this stage, let's ignore the cross validation and see what we mean in the framework of a regression problem. 

#### Workflow 

Every algorithm in scikit-learn generates an `Estimator` object associated to a specific machine learning algorithm. You then manipulate that object and use associated methods. Whatever the class (ML algorithm), the main methods are the same.  If you want to know more about the concepts / design driving the structure of scikit-learn, you may have a look at https://arxiv.org/abs/1309.0238.  

In general, the methodology for applying scikit-learn to a problem can be divided into several steps as described below: 

- 1. Choose a class of model by importing the appropriate estimator class from scikit-learn. I.e. you create an instance of an `estimator` object. All the **parameters** of an estimator can be set when it is instantiated, and have suitable default values.
- 2. Choose model hyperparameters by instantiating this class with desired values.
- 3. Arrange data into a features matrix and target vector following the discussion
from before.
- 4. Fit the model to your data by calling the `fit()` method of the model instance.
- 5. Apply the model to new data:
    * For supervised learning, often we predict labels for unknown data using the
`predict()` method.
    * For unsupervised learning, we often transform or infer properties of the data
using the `transform()` or `predict()` method.

**How to evaluate the quality of the prediction?** There is a score method associated to our instance. For this linear regression, the score returns $R^2$, namely the **coefficient of determination of Pearson**. Its mathematical expression is: 

$$ 
R^2 = 1 - \frac{SS_{\rm res}}{SS_{\rm tot}}   
$$

where $SS_{\rm res} = \sum_i^n (y_i - y_{i, \rm{mod}})^2$ is the sum of squares of the residuals (or residual sum of squares) and $SS_{\rm tot} = \sum_{i}^{n} (y_i - \bar{y})^2  $ (where $\bar{y}$ is the mean of the $y_i$ data) is the total sum of squares.

If the model values perfectly match the observed values, we have $R^2 = 1$. The model "explains" the variance in the data points (or equivalently, how much $R^2$ percents of the variations in $y$ is reduced by the variation in the "model" f(x)). A $R^2 = 1$ means that the model has a good predictive power. If $R^2 \sim 0$, it means that the model does not explain the distribution of values $y$. 

We'll come back at the score and at the interpretation of the quality of the model using test sets later.  

```python
# How to evaluate the quality of the prediction? 
model.score(X_test, y_test)
```

$$
P(L_k | \mathrm{features})

### Naive Bayes Classification

As the name states, this method uses the Bayes theorem, which describes the relationship of conditional probabilities of statistical quantities. 
Our data set is made of an ensemble of features (sepal and petal widths and lengths for the iris data) and of labels $L = {L_1, ..., L_n}$ (For the iris, we have 3 iris types corresponding to labels 1, 2, 3). 

Writing the Bayes theorem for a specific problem, we have: 

$$ 
P(L_k | \rm{features}) = \frac{P(\rm{features} | L_k) P(L_k) }{P(\rm{features})}
$$ 

If we are trying to decide between 2 labels (Let's say $L_1$ and $L_2$, then we can calculate the ratio of conditional probabilities: 

$$
\frac{P(L_1 | \rm{features})}{P(L_2 | \rm{features})} = \frac{P(\rm{features} | L_1) P(L_1) }{P(\rm{features} | L_2) P(L_2) } 
$$


Fine, but if we want to effectively calculate this, we need a model of $P(\rm{features} | L_k)$. Such a model is called *generative* model. It is called *generative* because it specifies the random process that generates the data. 

Finding this generative model from the training is the main piece of the training of Bayes classifiers. Here, we will make our life easy and *assume* that the data are described by a Gaussian distribution. At first, one can assume that there is conditional independance between the features. This is what is implement in the `GaussianNB` algorithm of sklearn. The next step is to relax this constrain on the conditional indepenance (see Chap. 9.3.3 of the astroML [book](#book)), allowing for covariances between the model distributions. This is however more demanding and effectively better handled with Gaussian mixture models that we will discuss later.

### Binary classification: True and False positives

For binary classification (2 labels), one generally assigns a **class 1** if the object is of the **type of interest**, and a 0 otherwise. In the exercise below, we will deal with an astronomical problem where we want to identify a certain class of stars, the so-called RR-Lyrae, among other stars. So the RR-Lyrae will have a class 1 and the other objects (stars) a class 0. 

Once our algorithm is trained, and we make our predictions for new data, four situations can occur as can be seen by building a confusion matrix as shown below: 

|N  | 1 (truth)| 0 (truth)  |   
|---|----|----|
| 1 (predicted)| True Positive = TP | False Positive = FP |
| 0 (predicted)| False Negative = FN | True Negative = TN | 


Specifically we have: 

- **True Positive (TP)** I have correctly assigned the label to my object. I say that I have an RR-Lyrae, and it is a RR-Lyrae. One speaks of a **True Positive (TP)**. 
- **True Negative (TN)** I yet have correctly rejected my object: it does not have the label of the class that I am "testing". I have an orange and I say that it is **not an apple**. 
- **[False Positive](https://en.wikipedia.org/wiki/False_positives_and_false_negatives#False_positive_error) (FP)**, where we have assigned a *true* class label when it is really false. We have classified as an apple but it is an orange ---> ***Type-1 error***.
* **[False Negative](https://en.wikipedia.org/wiki/False_positives_and_false_negatives#False_positive_error) (FN)**, where we have assigned a *false* class label when it is really true. We have an apple but we label it as an *not-apple* ---> ***Type-2 error***.

![](../Figures/False_True_positive_sketch.png)

(c) [Wikipedia](https://en.wikipedia.org/wiki/File:Precisionrecall.svg)

Based on these, we usually define several quantities that provide you different insights on the quality of the classification. In astronomy, one is often interested in the following 2 quantities: 

- The **Completeness** or **True Positive Rate (TPR)** (also sometimes called **sensitivity** or **recall**. The TPR is defined as the Fraction of positive instances correctly identified: $TPR = \frac{\rm TP}{\rm TP + FN}$. 
- The **Contamination** of **False DISCOVERY rate** is Fraction of positive instances incorrectly identified (i.e. among all positive instances): $\rm{FDR} = \frac{\rm FP}{\rm TP + FP}$
